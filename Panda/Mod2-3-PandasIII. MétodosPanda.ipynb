**2.7** Sacad los valores Ãºnicos y su frecuencia de las variables numÃ©ricas. Almacenad los resultados en un diccionario. Uno para cada variable.

```python
# Identificamos las variables numÃ©ricas
seleccion = df.select_dtypes(include = ["int16", "int32", "int64", "float16", "float32", "float64"])
seleccion.head()
```

```python
# Creamos tres diccionarios, ya que existen tres variables numÃ©ricas
dicc2= {}
dicc3= {}
dicc4= {}
```

```python
# Con un for vamos aÃ±adiendo a los diccionarios las variables y sus valores.
for col in df.columns:
    if col == "unnamed:_0":
        dicc2.setdefault(col, df[col].value_counts())
    elif col == "year":
        dicc3.setdefault(col, df[col].value_counts())
    elif col == "original_order":
        dicc4.setdefault(col, df[col].value_counts())
```
