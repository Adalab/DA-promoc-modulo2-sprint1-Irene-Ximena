```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_ataques = pd.read_csv("../Panda/5.pandas_mergeado.csv", index_col=0)
df_ataques.head(2)
```

1. Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.


```python
df_ataques_2015 = df_ataques[df_ataques['year'] >= 2015]
df_ataques_2015.tail(2)
```

```python
plt.figure(figsize = (15,10))

fig = sns.countplot( x= 'year', hue = 'activity', data = df_ataques_2015)
fig.set(xlabel = 'año', ylabel = 'número de casos')

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title = 'Actividades') #con este código especificamos que queremos la leyenda a la izquierda
```

Para el análisis de los datos hacemos unas comprobaciones:

```python
df_ataques_2010 = df_ataques[df_ataques['year'] >= 2010]
```

```python
df_analisis= pd.DataFrame(df_ataques_2010.groupby("year")['case_number'].count())
df_analisis.reset_index(inplace=True)
df_analisis.head(10)
```

Este gráfico nos muestra que en 2015 hay un máximo de ataques de tiburón, la mayoría producidos mientras se practicaban actividades deportivas acuáticas como el buceo o el surf. Este año marcó un récord de ataques de la pasada década, con un más de 30 ataques que en 2010, que lideraba la clasificacion hasta el momento. Los dos factores determinantes para este pico, son 'El Niño', un fenómeno meteorológico relacionado con el calentamineto del Pacífico y el aumento de las temperaturas derivada de la crisis climática. Como consecuencia de ambos, el habitat natural de estos animales se expandió. Así mismo, las temperaturas cálidas alargaron la temporada de baño, aumentando los encuentros enre personas y tiburones.


2. Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015.


```python
plt.figure(figsize = (10,8))

fig = sns.countplot( x= 'year', hue = 'sex_', data = df_ataques_2015)
fig.set(xlabel = 'año', ylabel = 'número de casos')
fig.set_title('Número de casos por sexo y año')

plt.legend(loc='upper right', borderaxespad=0, title = 'Sexo')
plt.show()
```

Como señalábamos en la anterior gráfica, el pico de ataques tuvo lugar en 2015 debido al aumento de temperaturas. En esta visualización comprobamos que los hombres, del 2015 al 2018, han sido los más damnificados por este animal. El fenómeno parece deberse a que los deportes acuáticos son actividades muy masculinazadas. Para hacernos una ida, según la Comisión de Deportes Australina, el 72 por ciento de los surferos en 2020 fueron hombres.

3. ¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.


```python
df_ataques_1500 = df_ataques[df_ataques['year'] > 1500]
df_ataques.head(2)
```

```python
plt.figure(figsize = (15,10))

fig = sns.histplot( x= 'year', data = df_ataques_1500, bins = 20)
fig.set(xlabel = 'año', ylabel = 'número de ataques')
fig.set_title('Número de ataques por año')

plt.show()
```

El mayor número de ataques se concentra de la década de 1900 en adelante, con un importante pico a partir del año 2000, donde encontramos más de 800 casos. El ascenso del número de casos desde el inicio del siglo pasado (con menos de 100 ataques identificados) se puede deber a varios factores: el primero, una mayor facilidad para el registro de estos sucesos debido a la cercanía temporal; también la creciente expansión y popularidad de los deportes acuáticos (por ejemplo, el paddle surf no surgió hasta la década de 1960) y el aumento del turismo, que a lo largo de las últimas décadas ha atraído a una población humanda en aumento a la costa.


4. Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
df_paises = df_ataques[((df_ataques['country'] == 'USA')| (df_ataques['country'] == 'AUSTRALIA') | (df_ataques['country'] == 'SOUTH AFRICA')) & (df_ataques['year'] > 1900)]
df_paises.head(2)
```

```python
df_usa_au_sa = pd.DataFrame(df_paises.groupby(["year", "country"])['case_number'].size()) #También podríamos utilizar .count()
df_usa_au_sa.reset_index(inplace=True)
df_usa_au_sa.head(10)
```

```python

plt.figure(figsize =(8,4)) # establecemos el marco de nuestra gráfica

fig = sns.lineplot(x='year' , y='case_number' , data = df_usa_au_sa, ci=None, hue ='country')
fig.set(xlabel='Año', ylabel = 'Número de ataques') # cambiamos los nombres de los ejes
fig.set_title('Evolución del número de ataques por año y país') # ponemos título a la gráfica
 
plt.legend(bbox_to_anchor=(1.05, 1), title = 'Países') #cambiamos la posición de la leyenda usando el parámetro bbox_to_anchor

plt.show() # mostramos la gráfica
```

Esto sería otra opción, pero es menos legible:

```python
plt.figure(figsize = (60,10))

fig = sns.countplot( x= 'year', hue = 'country', data = df_paises)
fig.set(xlabel = 'año', ylabel = 'número de casos')
fig.set_title('Número de casos por país y año')

plt.legend(loc='upper left', borderaxespad=0, fontsize=20, title = 'Países', title_fontsize=20) #cambiamos la posición de la leyenda usando el parámetro bbox_to_anchor
```
