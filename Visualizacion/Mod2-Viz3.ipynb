```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_ataques = pd.read_csv("../Panda/5.pandas_mergeado.csv", index_col=0)
df_ataques.head(2)
```

1. Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.


```python
df_ataques_2015 = df_ataques[df_ataques['year'] >= 2015]
df_ataques_2015.tail(2)
```

```python
plt.figure(figsize = (8,4))
fig = sns.countplot( x= 'year', hue = 'activity', data = df_ataques_2015)
fig.set(xlabel = 'año', ylabel = 'número de casos')
fig.set_title('Número de casos por actividad y año')
plt.show()
```

2. Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015.


```python
plt.figure(figsize = (8,4))
fig = sns.countplot( x= 'year', hue = 'sex_', data = df_ataques_2015)
fig.set(xlabel = 'año', ylabel = 'número de casos')
fig.set_title('Número de casos por sexo y año')
plt.show()
```

3. ¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.


```python
df_ataques_1500 = df_ataques[df_ataques['year'] > 1500]
df_ataques.head(2)
```

```python
plt.figure(figsize = (8,4))
fig = sns.histplot( x= 'year', data = df_ataques_1500, bins = 15)
fig.set(xlabel = 'año', ylabel = 'número de ataques')
fig.set_title('Número de ataques por año')
plt.show()
```

4. Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
df_paises = df_ataques[((df_ataques['country'] == 'USA')| (df_ataques['country'] == 'AUSTRALIA') | (df_ataques['country'] == 'SOUTH AFRICA')) & (df_ataques['year'] > 1900)]
df_paises.head(2)
```

```python
df_paises['year'].unique()
```

```python
#groupby,year/country, y luego case_number.count
#hacer el dataframe
```

```python
df_usa_au_sa = pd.DataFrame(df_paises.groupby(["year", "country"])['case_number'].size()) #También podríamos utilizar .count()
df_usa_au_sa.reset_index(inplace=True)
df_usa_au_sa.head(10)
```

```python


plt.figure(figsize =(8,4)) # establecemos el marco de nuestra gráfica

fig = sns.lineplot(x='year' , y='case_number' , data = df_usa_au_sa, ci=None, hue ='country')

fig.set(xlabel='Año', ylabel = 'Número de ataques') # cambiamos los nombres de los ejes

fig.set_title('Evolución del número de ataques por año y país') # ponemos título a la gráfica
 
plt.legend(bbox_to_anchor=(1.05, 1)) #cambiamos la posición de la leyenda usando el parámetro bbox_to_anchor

plt.show() # mostramos la gráfica
```

Esto sería otra opción, pero es menos legible:

```python
plt.figure(figsize = (16,8))
fig = sns.countplot( x= 'year', hue = 'country', data = df_paises)
fig.set(xlabel = 'año', ylabel = 'número de casos')
fig.set_title('Número de casos por país y año')
plt.show()
```

5. Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

```python
df_usa_au_sa.columns
```

```python
plt.figure(figsize = (16,8))

fig = sns.boxplot( x= 'country', y = 'case_number', data = df_usa_au_sa)

fig.set(xlabel = 'países', ylabel = 'número de casos')

fig.set_title('Datos estadísticos de los ataques por países')

plt.show()
```
